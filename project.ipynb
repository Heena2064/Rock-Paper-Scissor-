```python
import pandas as pd

df=pd.read_csv("Documents/smart_city_complaints 2.csv")
df.info()
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df['Department'].value_counts().plot(kind='bar',
color=['lightcoral'])

plt.title("Complaints by Department")
plt.xlabel("Department")
plt.ylabel("Number of Complaints")
plt.xticks(rotation=45)
plt.show()
```

```python
status_count = df['Status'].value_counts()


plt.figure(figsize=(6,6))
plt.pie(status_count,
labels=status_count.index,
autopct='%1.1f%%', startangle=140,
colors=['lightgreen','orange','red'])
plt.title("Complaint Status Distribution")
plt.axis('equal')
plt.show()
```

```python
dept_counts = df['Department'].value_counts()


plt.figure(figsize=(7,7))
plt.pie(dept_counts,
labels=dept_counts.index,
autopct='%1.1f%%', startangle=90,
colors=plt.cm.Paired.colors)
plt.title("Complaint by Department")
plt.axis('equal')
plt.show()
```

```python
severity_counts = df['Department'].value_counts()


plt.figure(figsize=(6,6))
plt.pie(severity_counts,
labels=severity_counts.index,
autopct='%1.1f%%', startangle=90,
colors=['gold','tomato','skyblue'])
plt.title("Complaint Severity Levels")
plt.axis('equal')
plt.show()
```
